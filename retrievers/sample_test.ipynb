```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse("sample.xml")
root = tree.getroot()
string = ET.tostring(root, encoding='unicode')
```

```python
tree1 = ET.fromstring(string)
```

```python
abstract_text = tree1.findall('.//ref')
citations = ["".join(element.itertext()) for element in abstract_text]
citations
```

```python
title = tree1.find('.//article-title')
title_text = "".join(title.itertext())
print(title_text)
```

```python
body = tree1.find('.//body')
body_text = " ".join(body.itertext())
print(body_text)
```
